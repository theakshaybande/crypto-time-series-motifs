# 01 Univariate Matrix Profile Basics

This notebook introduces the matrix profile on a simple univariate synthetic series. The main idea is that low profile values correspond to subsequences that have a close match elsewhere in the time series.

```python
from pathlib import Path
import sys

for candidate in [Path.cwd(), *Path.cwd().parents]:
    if candidate.name == "matrix_profile_lab":
        LAB_ROOT = candidate
        break
else:
    raise RuntimeError("Open this notebook from inside matrix_profile_lab.")

if str(LAB_ROOT) not in sys.path:
    sys.path.insert(0, str(LAB_ROOT))

from utils.data_generators import repeated_pattern_series
from utils.mp_helpers import compute_matrix_profile
from utils.plotting import plot_matrix_profile
```

```python
series, motif, motif_positions = repeated_pattern_series(
    length=260,
    motif_length=28,
    motif_positions=(35, 135, 205),
    noise=0.10,
    seed=11,
)

window = 28
mp = compute_matrix_profile(series, window=window)
plot_matrix_profile(
    series,
    mp["profile"],
    title="Univariate matrix profile: low values indicate repetition",
)

motif_positions
```
