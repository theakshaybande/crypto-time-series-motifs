# 02 Distance And Similarity

This notebook focuses on the distance profile. Instead of summarizing the whole series at once, it asks: for one query subsequence, how similar is every other subsequence? Comparing raw and z-normalized distances makes the role of scale and shape much clearer.

```python
from pathlib import Path
import sys

for candidate in [Path.cwd(), *Path.cwd().parents]:
    if candidate.name == "matrix_profile_lab":
        LAB_ROOT = candidate
        break
else:
    raise RuntimeError("Open this notebook from inside matrix_profile_lab.")

if str(LAB_ROOT) not in sys.path:
    sys.path.insert(0, str(LAB_ROOT))

import matplotlib.pyplot as plt

from utils.data_generators import repeated_pattern_series
from utils.mp_helpers import distance_profile_bruteforce
from utils.plotting import plot_distance_profile, plot_series_with_windows
```

```python
series, motif, motif_positions = repeated_pattern_series(
    length=260,
    motif_length=30,
    motif_positions=(50, 150, 210),
    noise=0.09,
    seed=9,
)

query_start = motif_positions[0]
window = len(motif)
raw_distance = distance_profile_bruteforce(series, query_start, window, normalize=False)
znorm_distance = distance_profile_bruteforce(series, query_start, window, normalize=True)

plot_series_with_windows(
    series,
    [(query_start, query_start + window)],
    title="Query subsequence used to build the distance profile",
    labels=["query"],
)

fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)
plot_distance_profile(raw_distance, title="Raw Euclidean distance profile", ax=axes[0])
plot_distance_profile(znorm_distance, title="Z-normalized distance profile", ax=axes[1])
fig.tight_layout()
```
