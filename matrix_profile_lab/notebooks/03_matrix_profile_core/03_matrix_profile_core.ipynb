# 03 Matrix Profile Core

This notebook connects the idea of a distance profile to the matrix profile. It computes a small brute-force matrix profile and compares it with the STUMPY result so the algorithmic summary becomes less abstract.

```python
from pathlib import Path
import sys

for candidate in [Path.cwd(), *Path.cwd().parents]:
    if candidate.name == "matrix_profile_lab":
        LAB_ROOT = candidate
        break
else:
    raise RuntimeError("Open this notebook from inside matrix_profile_lab.")

if str(LAB_ROOT) not in sys.path:
    sys.path.insert(0, str(LAB_ROOT))

import matplotlib.pyplot as plt

from utils.data_generators import repeated_pattern_series
from utils.mp_helpers import compute_bruteforce_matrix_profile, compute_matrix_profile
from utils.plotting import plot_matrix_profile
```

```python
series, motif, _ = repeated_pattern_series(
    length=180,
    motif_length=20,
    motif_positions=(25, 90, 140),
    noise=0.08,
    seed=4,
)

window = 20
bruteforce = compute_bruteforce_matrix_profile(series, window=window)
stumpy_mp = compute_matrix_profile(series, window=window)

plot_matrix_profile(series, stumpy_mp["profile"], title="Series and STUMPY matrix profile")

fig, ax = plt.subplots(figsize=(12, 3.2))
ax.plot(bruteforce["profile"], label="brute force", linewidth=2.0)
ax.plot(stumpy_mp["profile"], label="STUMPY", linewidth=2.0, linestyle="--")
ax.set_title("Brute-force and STUMPY profiles should closely match")
ax.set_xlabel("subsequence start")
ax.set_ylabel("profile")
ax.grid(alpha=0.25)
ax.legend(loc="best")
fig.tight_layout()
```
