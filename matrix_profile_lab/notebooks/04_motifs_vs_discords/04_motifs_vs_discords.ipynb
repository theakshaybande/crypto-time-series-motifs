# 04 Motifs Vs Discords

This notebook separates two core ideas: motifs are the most repeated subsequences, while discords are the most unusual subsequences. A single synthetic series can contain both, which makes the contrast very easy to see.

```python
from pathlib import Path
import sys

for candidate in [Path.cwd(), *Path.cwd().parents]:
    if candidate.name == "matrix_profile_lab":
        LAB_ROOT = candidate
        break
else:
    raise RuntimeError("Open this notebook from inside matrix_profile_lab.")

if str(LAB_ROOT) not in sys.path:
    sys.path.insert(0, str(LAB_ROOT))

import numpy as np

from utils.data_generators import repeated_pattern_series
from utils.mp_helpers import compute_matrix_profile, extract_top_discord, extract_top_motif
from utils.plotting import plot_motif_alignment, plot_series_with_windows
```

```python
series, motif, _ = repeated_pattern_series(
    length=280,
    motif_length=24,
    motif_positions=(40, 120, 210),
    noise=0.09,
    seed=14,
)

discord_start = 170
series[discord_start:discord_start + len(motif)] += np.linspace(-2.5, 2.5, len(motif))

mp = compute_matrix_profile(series, window=len(motif))
motif_a, motif_b = extract_top_motif(mp["profile"], mp["indices"])
discord = extract_top_discord(mp["profile"])

plot_series_with_windows(
    series,
    [
        (motif_a, motif_a + len(motif)),
        (motif_b, motif_b + len(motif)),
        (discord, discord + len(motif)),
    ],
    title="Motif pair versus discord in one synthetic series",
    labels=["motif A", "motif B", "discord"],
)

plot_motif_alignment(series, motif_a, motif_b, len(motif), title="Top motif pair overlay")

motif_a, motif_b, discord
```
