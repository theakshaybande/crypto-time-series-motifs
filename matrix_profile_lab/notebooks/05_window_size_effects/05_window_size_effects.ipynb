# 05 Window Size Effects

This notebook shows why window size matters so much. If the subsequence length is too short, the profile may match generic local wiggles. If it is too long, distinct patterns can be blurred together.

```python
from pathlib import Path
import sys

for candidate in [Path.cwd(), *Path.cwd().parents]:
    if candidate.name == "matrix_profile_lab":
        LAB_ROOT = candidate
        break
else:
    raise RuntimeError("Open this notebook from inside matrix_profile_lab.")

if str(LAB_ROOT) not in sys.path:
    sys.path.insert(0, str(LAB_ROOT))

import matplotlib.pyplot as plt

from utils.data_generators import repeated_pattern_series
from utils.mp_helpers import compute_matrix_profile
```

```python
series, motif, motif_positions = repeated_pattern_series(
    length=300,
    motif_length=26,
    motif_positions=(45, 150, 235),
    noise=0.10,
    seed=19,
)

window_sizes = [12, 26, 42]
fig, axes = plt.subplots(len(window_sizes) + 1, 1, figsize=(12, 9), sharex=False)
axes[0].plot(series, linewidth=2.0)
axes[0].set_title("Synthetic series used for the window-size comparison")
axes[0].grid(alpha=0.25)

for ax, window in zip(axes[1:], window_sizes):
    mp = compute_matrix_profile(series, window=window)
    ax.plot(mp["profile"], linewidth=2.0)
    ax.set_title(f"Matrix profile with window={window}")
    ax.set_ylabel("profile")
    ax.grid(alpha=0.25)

axes[-1].set_xlabel("subsequence start")
fig.tight_layout()

motif_positions
```
