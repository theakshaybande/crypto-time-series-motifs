# 06 Multivariate Intro

This notebook introduces multivariate matrix profile on a tiny three-channel synthetic series. The point is not complexity yet. It is to see how a shared event across channels can become easier to detect when dimensions are analyzed together.

```python
from pathlib import Path
import sys

for candidate in [Path.cwd(), *Path.cwd().parents]:
    if candidate.name == "matrix_profile_lab":
        LAB_ROOT = candidate
        break
else:
    raise RuntimeError("Open this notebook from inside matrix_profile_lab.")

if str(LAB_ROOT) not in sys.path:
    sys.path.insert(0, str(LAB_ROOT))

import matplotlib.pyplot as plt

from utils.data_generators import multivariate_motif_series
from utils.mp_helpers import multivariate_matrix_profile
from utils.plotting import plot_multivariate
```

```python
channels, motif_positions = multivariate_motif_series(
    length=320,
    motif_length=28,
    motif_positions=(70, 210),
    noise=0.08,
    seed=21,
)

window = 28
mp = multivariate_matrix_profile(channels, window=window)
mean_profile = mp["profile"].mean(axis=0)

plot_multivariate(channels, labels=["channel A", "channel B", "channel C"], title="Tiny multivariate motif example")

fig, ax = plt.subplots(figsize=(12, 3.2))
ax.plot(mean_profile, linewidth=2.0)
ax.set_title("Mean multivariate profile across channels")
ax.set_xlabel("subsequence start")
ax.set_ylabel("profile")
ax.grid(alpha=0.25)
fig.tight_layout()

motif_positions
```
