# 09 Visualization

This notebook focuses on visual storytelling. The exact matrix profile numbers matter, but the way motifs and discords are visualized often determines whether the result is understandable.

```python
from pathlib import Path
import sys

for candidate in [Path.cwd(), *Path.cwd().parents]:
    if candidate.name == "matrix_profile_lab":
        LAB_ROOT = candidate
        break
else:
    raise RuntimeError("Open this notebook from inside matrix_profile_lab.")

if str(LAB_ROOT) not in sys.path:
    sys.path.insert(0, str(LAB_ROOT))

import numpy as np

from utils.data_generators import repeated_pattern_series
from utils.mp_helpers import compute_matrix_profile, extract_top_discord, extract_top_motif
from utils.plotting import plot_matrix_profile, plot_motif_alignment, plot_series_with_windows
```

```python
series, motif, _ = repeated_pattern_series(
    length=300,
    motif_length=26,
    motif_positions=(55, 160, 235),
    noise=0.10,
    seed=41,
)

series[115:141] += np.r_[np.linspace(0.0, 2.8, 13), np.linspace(2.8, -1.2, 13)]

window = 26
mp = compute_matrix_profile(series, window=window)
motif_a, motif_b = extract_top_motif(mp["profile"], mp["indices"])
discord = extract_top_discord(mp["profile"])

plot_series_with_windows(
    series,
    [(motif_a, motif_a + window), (motif_b, motif_b + window), (discord, discord + window)],
    title="Annotated view: motif pair and discord",
    labels=["motif A", "motif B", "discord"],
)

plot_motif_alignment(series, motif_a, motif_b, window, title="Aligned motif view")
plot_matrix_profile(series, mp["profile"], title="Profile view for the same series")

motif_a, motif_b, discord
```
